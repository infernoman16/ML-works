```
df.drop(['pdate','pcircle','msisdn'],axis=1,inplace=True)
```

```
df1=df.copy()
```

```
df1.drop('label',axis=1,inplace=True)
```

```
from sklearn.decomposition import PCA
```

```
pca=PCA()
```

```
pca.fit(df1)
```

```
data_p=pca.transform(df1)
```

```
pca=PCA(n_components=13)
pca.fit(df1)
data_p=pd.DataFrame(pca.transform(df1))
```

```
data_p.shape
```

```
X1=data_p
y1=df['label']
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.3, random_state=1)
```

```
from sklearn.ensemble import RandomForestClassifier
```

```
rfc=RandomForestClassifier(n_estimators=300)
```

```
rfc.fit(X_train,y_train)
```

```
prediction=rfc.predict(X_test)
```

```
from sklearn.metrics import classification_report,confusion_matrix
```

```
print(classification_report(y_test,prediction))
```

```
pred_prob=rfc.predict_proba(X_test)
```

```
pred_df=pd.DataFrame(pred_prob)
```

```
pred_df.drop(pred_df.columns[0],axis=1,inplace=True)
```

```
y=pd.DataFrame(y_test,index=None)
```

```
y.index=np.arange(0,len(y))
```

```
frames=[pred_df,y]
```

```
final_df=pd.concat(frames,axis=1)
```
